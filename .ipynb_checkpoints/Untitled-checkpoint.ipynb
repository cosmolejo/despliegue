```python
race = ['Indian',
'Latino',
'Middle Eastern',
'White',
'Middle Eastern',
'White',
'Black',
'Asian',
'Indian',
'Latino',
'Asian',
'Black']
```

```python
import matplotlib.pyplot as plt

lista = np.loadtxt('./Sketch/lista', dtype='str')[:-1]

for f in zip(lista,race):
    im = Image.open(f"./Sketch/{f}")
    res = generator.generate_face(im)
    res = Image.fromarray(res)
    res.save(f'./FaceGenRace_1/{f}')
    plt.subplot(121),plt.imshow(im, cmap='gray')
    plt.title('sketch')
    plt.subplot(122),plt.imshow(res)
    plt.title('Rostro generado')
    
    plt.show()
```
